```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
train_path = 'gdrive/My Drive/Genre Classification Dataset/train_data.txt'
test_path = 'gdrive/My Drive/Genre Classification Dataset/test_data.txt'
```

```python
df = pd.read_csv(train_path, sep=':::',engine = 'python',names=['Title', 'Genre', 'Description'])
df
```

```python
df.describe()
```

```python
x_train = df.drop('Genre', axis='columns')
y_train = df['Genre']
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(max_features=5000)
x_train_trans = tfidf_vectorizer.fit_transform(x_train['Description'])
x_train_trans
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
```

```python
model.fit(x_train_trans, y_train)
```

```python
y_train_pred = model.predict(x_train_trans)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print("Accuracy :", accuracy_score(y_train, y_train_pred))
```

```python
df1 = pd.read_csv(test_path, sep=':::',engine = 'python',names=['Title', 'Genre', 'Description'])
df1
```

```python
x_test = df1.drop('Genre', axis='columns')
y_test = df1['Genre']
```

```python
X_test_trans = tfidf_vectorizer.fit_transform(x_test['Description'])
x_test.shape
```

```python
X_test_pred = model.predict(X_test_trans)
df1['Predicted_Genre'] = X_test_pred
```

```python
df1.to_csv('predicted_genres.csv', index=False)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, X_test_pred)
cm
```
